```python

#How many times has it taken more than 5 days from placing an order to shipping
data1['Ship Date']=pd.to_datetime(data1['Ship Date'])
data1['Order Date']=pd.to_datetime(data1['Order Date'])

time = data1['Ship Date']- data1['Order Date']
sum(time >'5 days')
```

```python
#Find the total number of orders in every category which has been shipped with a duration > 5 days
data1['Order Date'] =pd.to_datetime(data1['Order Date'])
data1['Ship Date'] = pd.to_datetime(data1['Ship Date'])
data1['Timetaken'] = data1['Ship Date']- data1['Order Date']
kk = data1[data1.Timetaken > '5 days']
kk.groupby('Category').count()['Timetaken']
```

```python
#What’s the percentage of items which has been shipped within 5 days
data1['Order Date'] =pd.to_datetime(data1['Order Date'])
data1['Ship Date'] = pd.to_datetime(data1['Ship Date'])
data1['Timetaken'] = data1['Ship Date']- data1['Order Date']
a = data1[data1.Timetaken < '5 days'].count()['Timetaken']
b = data1.count()['Timetaken']
perc = a/b*100
print(perc)
```

```python
#What’s the percentage of items which has been shipped after 5 days
data1['Order Date'] =pd.to_datetime(data1['Order Date'])
data1['Ship Date'] = pd.to_datetime(data1['Ship Date'])
data1['Timetaken'] = data1['Ship Date']- data1['Order Date']

a = data1[data1.Timetaken > '5 days'].count()['Timetaken']
b = data1.count()['Timetaken']
perc = a/b*100
print(perc)
```
